```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('psi_df_2016_2019.csv')
# df.head(5)
```

```python
x = ['national', 'south', 'north', 'east', 'central'] 
list_ = [df]
for psi_df in list_:
    df['mean']=df[x].mean(axis=1, skipna=True)
    df['median']=df[x].median(axis=1, skipna=True)
    
df['year'] = pd.DatetimeIndex(df['timestamp']).year
df['month'] = pd.DatetimeIndex(df['timestamp']).month
```

```python
for i in x:
    df.loc[df[i] <= 55, 'cat '+ i] = 0 #0 for Normal
    df.loc[(df[i] > 55 ) & (df[i] <= 150) , 'cat '+ i] = 1 #1 for Elevated
    df.loc[(df[i] > 150) & (df[i] <= 250) , 'cat '+ i] = 2 #2 for High
    df.loc[df[i] > 250 , 'cat '+ i] = 3 #3 for Very High

df.head()
```

```python
feature_names = [ 'year', 'month', 'mean', 'median']
target_names = ['cat national', 'cat south', 'cat north', 'cat east', 'cat central']
X = df[feature_names]
Y = df[target_names]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
score = model.score(x_test, y_test)
print('Score', score)
pre = model.predict(x_test)
```

```python
pre_df = pd.DataFrame(pre, columns = target_names)
pre_df.head(5)
```
